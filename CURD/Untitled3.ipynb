```python
from tkinter import *
import sqlite3
import tkinter.ttk as ttk
import tkinter.messagebox as tkMessageBox

root = Tk()
root.title('BOOK MANAGEMENT SYSTEM APPLICATION')

def Database():
    global con,cursor
    con = sqlite3.connect('book1.db')
    cursor = con.cursor()
    cursor.execute("CREATE TABLE IF NOT EXISTS  member1(bookname TEXT, author TEXT, year TEXT, isbn TEXT, statusissue TEXT)")

def AddBook():
    if BOOKNAME.get() == "" or AUTHOR.get() == "" or YEAR.get() == "" or ISBN.get() == "":
        txt_result.config(text="Please enter all the fields",fg="red")
    else:
        Database()
        cursor.execute("INSERT INTO member1(bookname,author,year,isbn) VALUES(?,?,?,?)",(str(BOOKNAME.get()), str(AUTHOR.get()), str(YEAR.get()), str(ISBN.get())))
        con.commit()
        BOOKNAME.set("")
        AUTHOR.set("")
        YEAR.set("")
        ISBN.set("")
        cursor.close()                                                                                                                
        con.close()
        txt_result.config(text="ALL RECORD INSERTED !!",fg="green")
    
def Search():
    if not tree.selection():
        txt_result.config(text="Please select an item first", fg="red")
    else:
        result = tkMessageBox.askquestion('Do you want to Issue this Book ?(y/n)', icon="warning")
        if result == 'yes':
            curItem = tree.focus()
            contents =(tree.item(curItem))
            selecteditem = contents['values']
            tree.delete(*tree.get_children())
            con = sqlite3.connect('book.db')
            cur = con.cursor()
            cursor.execute("UPDATE member1 SET status=?",str(STATUS.get()))
            con.commit()
            STATUS.set("")
            con.close()
            txt_result.config(text="Sucessfully Changed Status of Book", fg="green")
            
def IssueBook():
    if not tree.selection():
        txt_result.config(text="Please select an item first", fg="red")
    else:
        result = tkMessageBox.askquestion('Do you want to Issue this Book ?(y/n)', icon="warning")
        if result == 'yes':
            curItem = tree.focus()
            contents =(tree.item(curItem))
            selecteditem = contents['values']
            tree.delete(*tree.get_children())
            con = sqlite3.connect('book1.db')
            cur = con.cursor()
            cursor.execute("UPDATE member1 SET statusissue=?",str(STATUSISSUE.get()))
            con.commit()
            STATUSISSUE.set("")
            con.close()
            txt_result.config(text="Sucessfully Changed Status of Book", fg="green")
    
def Read():
    tree.delete(*tree.get_children())
    Database()
    cursor.execute('SELECT * FROM member1')
    fetch = cursor.fetchall()
    for data in fetch:
        tree.insert('', 'end', values=(data[0],data[1],data[2],data[3],data[4]))
    cursor.close()
    txt_result.config(text="SUCESSFULLY FETCHED DATA",fg="black")    
    
    
def Delete():
    if not tree.selection():
        txt_result.config(text="Please select an item first", fg="red")
    else:
        result = tkMessageBox.askquestion('Do you want to Delete?(y/n)', icon="warning")
        if result == 'yes':
            curItem = tree.focus()
            contents =(tree.item(curItem))
            selecteditem = contents['values']
            tree.delete(curItem)
            Database()
            cursor.execute("DELETE FROM member1")
            con.commit()
            cursor.close()
            con.close()
            txt_result.config(text="Sucessfully Deletd data", fg="red")

def Exit():
    result = tkMessageBox.askquestion('Do you want to Exit?(y/n)', icon = "warning")
    if result == "yes":
        root.destroy()
        exit()
        

BOOKNAME = StringVar()
AUTHOR = StringVar()
YEAR = StringVar()
ISBN = StringVar()
STATUSISSUE = StringVar()

Top = Frame(root,width =900, height =500,bg="red" ,bd =8, relief = 'raise')
Top.pack(side=TOP)
Left = Frame(root,width =300, height =500,bg="green", bd =8, relief = 'raise')
Left.pack(side=LEFT)
Right = Frame(root,width =300, height =500,bg="blue", bd =8, relief = 'raise')
Right.pack(side=RIGHT)

Forms = Frame(Left, width =300, height =450)
Forms.pack(side=TOP)

Buttons = Frame(Left,width =300, height =100, bd =8, relief ='raise')
Buttons.pack(side=BOTTOM)

RadioGroup = Frame(Forms)
MarkedIssued=Radiobutton(RadioGroup,text='Issued',variable=STATUSISSUE,value='Issued').pack(side=LEFT)
MarkedNotIssed=Radiobutton(RadioGroup,text='Available',variable=STATUSISSUE,value='Available').pack(side=LEFT)


txt_title = Label(Top,width =900,font=('arial',24),bg="orange",text="BOOK MANAGEMENT SYSTEM APPLICATION")
txt_title.pack()
txt_bookname = Label(Forms,text="Bookname:",font=('arial',16),bd=15)
txt_bookname.grid(row=0,stick="e")
txt_author = Label(Forms,text="Author:",font=('arial',16),bd=15)
txt_author.grid(row=1,stick="e")
txt_year = Label(Forms,text="Year:",font=('arial',16),bd=15)
txt_year.grid(row=2,stick="e")
txt_isbn= Label(Forms,text="ISBN:",font=('arial',16),bd=15)
txt_isbn.grid(row=3,stick="e")
txt_issue = Label(Forms,text="Status:",font=('arial',16),bd=15)
txt_issue.grid(row=4,stick="e")
txt_result = Label(Buttons)
txt_result.pack(side=TOP)

bookname=Entry(Forms,textvariable=BOOKNAME,width=20)
bookname.grid(row=0,column=1)
author=Entry(Forms,textvariable=AUTHOR,width=20)
author.grid(row=1,column=1)
year=Entry(Forms,textvariable=YEAR,width=20)
year.grid(row=2,column=1)
isbn=Entry(Forms,textvariable=ISBN,width=20)
isbn.grid(row=3,column=1)
RadioGroup.grid(row=4,column=1)

btn_addbook =Button(Buttons,width=10,text="Add Book",command=AddBook)
btn_addbook.pack(side=LEFT)
btn_issuebook =Button(Buttons,width=10,text="Issue Book",command=IssueBook)
btn_issuebook.pack(side=LEFT)
btn_read =Button(Buttons,width=10,text="Read",command=Read)
btn_read.pack(side=LEFT)
btn_delete =Button(Buttons,width=10,text="Delete",command=Delete)
btn_delete.pack(side=LEFT)
btn_exit =Button(Buttons,width=10,text="Exit",command=Exit)
btn_exit.pack(side=LEFT)

scrollbary=Scrollbar(Right, orient=VERTICAL)
scrollbarx=Scrollbar(Right, orient=HORIZONTAL)

tree = ttk.Treeview(Right, columns=("Bookname","Author","Year","Isbn","Status"))

scrollbary.config(command=tree.yview)
scrollbary.pack(side=RIGHT,fill=Y)
scrollbarx.config(command=tree.xview)
scrollbarx.pack(side=BOTTOM,fill=X)
tree.heading('Bookname',text="Bookname",anchor=W)
tree.heading('Author',text="Author",anchor=W)
tree.heading('Year',text="Year",anchor=W)
tree.heading('Isbn',text="ISBN",anchor=W)
tree.heading('Status',text="Status",anchor=W)
tree.column('#0', stretch=NO,minwidth=NO,width=0)
tree.column('#1', stretch=NO,minwidth=NO,width=80)
tree.column('#2', stretch=NO,minwidth=NO,width=120)
tree.column('#3', stretch=NO,minwidth=NO,width=80)
tree.column('#4', stretch=NO,minwidth=NO,width=120)
tree.column('#5', stretch=NO,minwidth=NO,width=120)
tree.pack()

if __name__== '__main__':
    Read()
    root.mainloop()
    
    
```
