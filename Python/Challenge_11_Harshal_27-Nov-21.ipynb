```python
# Q-2. Create tkinter application to accept user name in a text widget. Then display the user name with a welcome message in a label widget

window = Tk()

def name():
    name = "Welcome "+str(e1_nam.get()) +"!!!"
    t1.delete("1.0",END)
    t1.insert(END,name)
    label.config(text=name)
    
l1 = Label(window,text="Enter the Name ")
l1.pack()

e1_nam = StringVar()
e1_nam = Entry(window,textvariable = e1_nam)
e1_nam.pack()

l2 = Label(window,text="Message")
l2.pack()

t1 = Text(window,height=1,width=20)
t1.pack()

b1 = Button(window,text="print",bg="yellow",height=1,width=10,command=name)
b1.pack()

label = Label(window)

window.mainloop()
```

```python
# Q-7. Create a tkinter application to display a menu with three items. A separate message should be display when each of the three menu items is clicked.

window = Tk()

def pyt():
    print("PYTHON")
    
def jav():
    print("JAVA")
    
def cp():
    print("C++")
    
menubar = Menu(window)

filemenu = Menu(menubar)

filemenu.add_command(label="Python",command = pyt)
filemenu.add_separator()

filemenu.add_command(label="Java",command = jav)
filemenu.add_separator()

filemenu.add_command(label="C++",command = cp)
filemenu.add_separator()

menubar.add_cascade(label="File",menu=filemenu)

window.config(menu=menubar)

window.mainloop()
```

```python
# Q-8. Create a tkinter application to display a radio button group with three items. A separate message should be display when each of the three radio buttons items is clicked.

window = Tk()

def gen():
    gender = "You are "+var.get()
    label.config(text=gender)
    
var = StringVar()

R1 = Radiobutton(window,text="Male",variable=var,value="Male",command=gen)
R1.pack()

R2 = Radiobutton(window,text="Female",variable=var,value="Female",command=gen)
R2.pack()

R3 = Radiobutton(window,text="Transgender",variable=var,value="Transgender",command=gen)
R3.pack()

label= Label(window)
label.pack()

window.mainloop()
```

```python
# Q-9. Create a tkinter application to display three check boxes. A separate message should be display when each of the three check boxes items is selected.

def sel1():
    if CheckVar1.get() == 1:
        selection = "You have selected Bike Option"
    else:
        selection = "You have de-selected the Bike option"
        
    label.config(text=selection)
    
def sel2():
    if CheckVar2.get() == 1:
        selection = "You have selected Car Option"
    else:
        selection = "You have de-selected the Car option"
    label.config(text=selection)
        
def sel3():
    if CheckVar3.get() == 1:
        selection = "You have selected Train Option "
    else:
        selection = "You have de-selected the Train option"
        
    label.config(text=selection)
```

```python
window = Tk()

CheckVar1 = IntVar()
CheckVar2 = IntVar()
CheckVar3 = IntVar()

c1 = Checkbutton(window,text="Bike Option",variable=CheckVar1,onvalue=1,offvalue=0,height=5,width=20,command=sel1)
c2 = Checkbutton(window,text="Car Option",variable=CheckVar2,onvalue=1,offvalue=0,height=5,width=20,command=sel2)
c3 = Checkbutton(window,text="Train Option",variable=CheckVar3,onvalue=1,offvalue=0,height=5,width=20,command=sel3)
c1.pack()
c2.pack()
c3.pack()

label = Label(window)
label.pack()

window.mainloop()
```
