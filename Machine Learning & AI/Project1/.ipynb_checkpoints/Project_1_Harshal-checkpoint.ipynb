### Modify the Survival Status column value to be meaningfull as categorised

```python
stock_1_data['Year'] = stock_1_data['Year'].map({2008:"A",2009:"B",2010:"C",2011:"D",2012:"E",2013:"F",2014:"G",2015:"H",2016:"I",2017:"J",2018:"K",2019:"L",2020:"M",2021:"N"})
stock_1_data['Year'] = stock_1_data['Year'].astype('category')
print(stock_1_data.head)
```

```python
print('Shape: ',stock_1_data.shape)
print('Column Name:',stock_1_data.columns)
print('Unique :',stock_1_data['Year'].unique())
print('Value Count:',stock_1_data['Year'].value_counts())
```

## Histogram,Pdf & Cdf

```python
A =stock_1_data.loc[stock_1_data['Year']=="A" ]
B =stock_1_data.loc[stock_1_data['Year']=="B"]
C =stock_1_data.loc[stock_1_data['Year']=="C"]
D =stock_1_data.loc[stock_1_data['Year']=="D"]
E =stock_1_data.loc[stock_1_data['Year']=="E"]
plt.figure(1,figsize=(20,7))

plt.subplot(121)
plt.plot(A['Volume'],np.zeros_like(A['Volume']),'o',label='2008')
plt.plot(B['Volume'],np.zeros_like(B['Volume']),'o',label='2009')
plt.plot(C['Volume'],np.zeros_like(C['Volume']),'o',label='2010')
plt.plot(D['Volume'],np.zeros_like(D['Volume']),'o',label='2011')
plt.plot(E['Volume'],np.zeros_like(E['Volume']),'o',label='2012')
plt.legend()
plt.xlabel('Volume')
plt.title('Histogram on the basis of Volume For FY2008-12')


plt.subplot(122)
plt.plot(A['Close'],np.zeros_like(A['Close']),'o',label='2008')
plt.plot(B['Close'],np.zeros_like(B['Close']),'o',label='2009')
plt.plot(C['Close'],np.zeros_like(C['Close']),'o',label='2010')
plt.plot(D['Close'],np.zeros_like(D['Close']),'o',label='2011')
plt.plot(E['Close'],np.zeros_like(E['Close']),'o',label='2012')
plt.legend()
plt.xlabel('Close')
plt.title('Histogram on the basis of Close For FY2008-12')
plt.show()
```

```python
sns.FacetGrid(stock_1_data, hue='Year',height=7).map(sns.histplot,'Volume').add_legend()
plt.title('Histogram for Volume')
plt.show()
```

```python
sns.FacetGrid(stock_1_data, hue='Year',height=7).map(sns.histplot,'Close').add_legend()
plt.title('Histogram for Closing Stock')
plt.show()
```

```python
feature
```

```python
plt.figure(3,figsize=(20,5))
for idx, feature in enumerate(list(A.columns)[1:5]):
    plt.subplot(1,5,idx+1)
    print('-'*30+feature+'-'*30)
    print(feature)
    counts, bin_edges = np.histogram(A[feature], bins=10, density=True)
    print("Bin_Edges: {}".format(bin_edges))
    pdf = counts/sum(counts)
    print("PDF:{}".format(pdf))
    cdf = np.cumsum(pdf)
    print("CDF :{}".format(cdf))
    plt.plot(bin_edges[1:],pdf,label ='pdf_A')
    plt.plot(bin_edges[1:],cdf,label ='cdf_A')
    
    print('-'*30+"B"+'-'*30)
    print(feature)
    counts, bin_edges = np.histogram(B[feature], bins=10, density=True)
    print("Bin_Edges: {}".format(bin_edges))
    pdf = counts/sum(counts)
    print("PDF:{}".format(pdf))
    cdf = np.cumsum(pdf)
    print("CDF :{}".format(cdf))
    plt.plot(bin_edges[1:],pdf,label ='pdf_B')
    plt.plot(bin_edges[1:],cdf,label ='cdf_B')
    
    plt.title('pdfand cdf based on ' +feature)
    plt.legend()
    plt.xlabel(feature)
```
